##### Veient la sortida del mètode _describe_, s'observa que el valor mínim d'_CRSElapsedTime_ és negatiu. Aquest valor és erròni: un retard pot ser negatiu si s'arriba abans d'hora, però el valor d'una durada només pot ser positiu. Per veure la magnitud d'aquest problema, s'endreça al Data Frame en aquesta variable: només hi ha un registre afectat per aquest error, per tant, es pot eliminar sense més efectes. 

```python
df = df.sort_values(by=['CRSElapsedTime'])[1:df.shape[0]]
df.describe()
```

##### Ja no apareixen durades mínimes negatives.
##### *(3)* Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

```python
# La velocitat mitjana del vol és la distancia recorreguda dividida per la durada real del vol
df["VelocitatMitjana"]= df["Distance"] / df["AirTime"]
# Es pot saber si el vol ha arribat tard comparant la durada real amb la prevista.
df["ElapsedTimeDelay"] = df["ActualElapsedTime"]-df["CRSElapsedTime"]
# Es pot calcular l'endarreriment total del vol restant la durada prevista a la real. 
```

```python
df.head()
```

##### He revisat la definició de les variables _ActualElapsedTime_ i _CRSElapsedTime_ perquè trobo que hi ha una diferència molt gran en alguns casos i dubto que hi hagi vols amb durades < 10 minuts, però sembla que és correcte. En tot cas, potser hi ha valors anòmals que no són ni negatius ni nuls.

##### (4) Taula de les aerolínies amb més endarreriments acumulats.
##### D'entrada, cal considerar que no tenim les dades d'aerolínia, però com que els avions sempre volen amb la mateixa companyia, podem fer l'anàlisi per l'identificador d'avió enlloc d'això. 
##### En primer lloc podríem veure quins avions han obtingut els endarreriments més grans:

```python
endarreriments = df[['TailNum','ElapsedTimeDelay']]
endarreriments.sort_values(by=['ElapsedTimeDelay'],ascending=False)[0:10]
```

##### Per calcular l'endarreriment acumulat, hauriem de sumar els endarreriments que ha fet cada avió en tots els seus trajectes:

```python
suma_endarreriments = endarreriments.groupby('TailNum').ElapsedTimeDelay.sum()
suma_endarreriments.sort_values(ascending = False)[:10]
```

##### Com que segurament no hi ha el mateix nombre de registres per cada avió, és més adequat calcular la mitjana; el procediment és molt semblant:

```python
mitjana_endarreriments = endarreriments.groupby('TailNum').ElapsedTimeDelay.mean()
mitjana_endarreriments.sort_values(ascending = False)[:10]
```

##### (5) Quins són els vols més llargs? I els més endarrerits?
##### Els vols es designen pel _FlightNum_, o sigui que cal treballar amb aquesta variable. Per estudiar la durada dels vols, agafarem la durada teòrica; i pel que fa als endarreriments, la varible que hem calculat.

```python
vols = df[['FlightNum','CRSElapsedTime','ElapsedTimeDelay']]
vols_llargs = vols.groupby('FlightNum').CRSElapsedTime.max()
vols_llargs.sort_values(ascending = False)[:10]
```

```python
vols_endarrerits = vols.groupby('FlightNum').ElapsedTimeDelay.max()
vols_endarrerits.sort_values(ascending = False)[:10]
```
